```python
import gradio as gr

def make_prediction(*input_features):
    try:
        number = [list(map(float, input_features))]
        
        # Predict using your model (e.g., RFC_model)
        prediction = RFC_model.predict(number)

        
        info = ''
        if prediction[0] == 0:
            info = 'Hazardous'
        elif prediction[0] == 1:
            info = 'Poor'
        elif prediction[0] == 2:
            info = 'Moderate'
        else:
            info = 'Good'
        
        final_info = f"The air quality is: {info}"
        return final_info
    except ValueError:
        return "Error: Please enter valid inputs."

headline = 'Air Quality Check'


feature_labels = [
    "Temperature", 
    "Humidity", 
    "NO2", 
    "SO2", 
    "CO", 
    "Proximity to Industrial Areas", 
    "Population Density"
]


inputs = [gr.Number(label=label) for label in feature_labels]


iface = gr.Interface(
    fn=make_prediction,
    inputs=inputs,
    outputs='text',
    title=headline
)

iface.launch(share=True)
```
